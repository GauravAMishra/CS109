```python
# special IPython command to prepare the notebook for matplotlib
%matplotlib inline 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/home/webwerks/Salaries.csv")
teamIDs = np.unique(df['teamID'].values)
teamIDs
```

```python
salary_sum = {}
for team in teamIDs:
    salary_sum[team] = df[df['teamID']==team]['salary'].sum()
    
salary_sum
```

```python
#Problem 1(b)

#Add the total sum of salary in data frame
df['sum_salary']= df.groupby(['yearID','teamID'])['salary'].transform(np.sum)
```

```python
#Remove the unneccesary column
df = df[['yearID','teamID','sum_salary']] #df.drop(['lgID','playerID','salary'], axis=1)
df = df.drop_duplicates(subset=['yearID','teamID'])
```

```python
#sort by team and year and problem solved
df = df.sort_values(['teamID','yearID'])
df.head()
```

```python
df2 = pd.read_csv("/home/webwerks/Teams.csv")
df2.head()
```

```python
df2['sum_wins'] = df2.groupby(['yearID','teamID'])['W'].transform(np.sum)
df2.head()
```

```python
#Problem 1(c)

#get data frame
df2 = df2[['yearID','teamID','sum_wins']]
df2 = df2.sort_values(['teamID','yearID'])
df2.head()
```

```python
#merge two dataframe
total_rows = pd.merge(df,df2,how='outer',on=['yearID','teamID'])
```
